# Que-1) Find all records from data when the weather was exactly clear?

```python
# There are three ways this question can be approached 
# 1. By filtering the dataset - extracting the exact info from rows that matches the condition

data[data['Weather']=='Clear'] # displays whole dataframe
```

```python
# 1.1 - to show only records
data[data['Weather']=='Clear'].shape
```

```python
# Using Grouping 
# Group By Command group the data according to each unique value and we can use the aggregate function on that to get the desired number of rows which has weather as clear.

data.groupby('Weather').get_group('Clear')
```

```python
# displays shape
data.groupby('Weather').get_group('Clear').shape
```

# Que-6) Find the number of records where wind speed is greater than 24 and visibility is equal to 25?

```python
 # Finding the number of recordes, where wind speed is greater than 24 and visibility is equal to 25

data[(data['Wind Speed_km/h']>24) & (data['Visibility_km']==25)]
```

```python
# finding the number of records
data[(data['Wind Speed_km/h']>24 & (data['Visibility_km']==25))].shape
```

# Que-7) What is the mean value of each column against each weather condition?

```python
# whenever the question asks for each value - use the groupby() function

data.groupby('Weather_Condition').mean().shape
```

# Que-8) Find all instances where weather is clear and relative humidity is greater than 50 or visibility is above 40?

```python
 # This kind of question can be solved using filtering method
    
data[(data['Weather_Condition'] == 'Clear') & ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))]
```

```python
 # Finding the number of records

data[(data['Weather_Condition'] == 'Clear') & ((data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40))].shape
    
```
